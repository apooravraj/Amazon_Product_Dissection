# 8. Database Relationships

A relational database works by connecting different tables through Primary Keys (PK) and Foreign Keys (FK). These relationships help maintain data integrity and reduce redundancy.

The Amazon database consists of three major types of relationships:

### 1. One-to-One (1:1)

A one-to-one relationship exists when one record in a table is associated with only one record in another table.

**Example:**

- One Order → One Payment

Every order has one payment record associated with it.

---

### 2. One-to-Many (1:M)

A one-to-many relationship exists when one record in a table can be associated with multiple records in another table.

Examples:

- One User → Many Orders
- One User → Many Addresses
- One Seller → Many Products
- One Category → Many Products
- One Product → Many Reviews
- One Cart → Many Cart Items
- One Order → Many Order Items

These relationships help organize data efficiently without duplication.

---

### 3. Many-to-Many (M:N)

A many-to-many relationship exists when multiple records from one table are related to multiple records in another table.

Examples:

- Users ↔ Products (through Wishlist)
- Carts ↔ Products (through Cart_Items)
- Orders ↔ Products (through Order_Items)

These relationships are implemented using bridge tables.

## Relationship Summary

# 9. Database Design Rationale

The Amazon database has been designed using the principles of relational database management systems (RDBMS).

The following design decisions were considered while creating the schema:

- Each table represents a single business entity.
- Every table contains a Primary Key to uniquely identify each record.
- Foreign Keys establish relationships between related tables.
- Duplicate data is minimized through normalization.
- Order details are separated from product information to avoid redundancy.
- Cart items and Order items are maintained in separate tables because a single cart or order can contain multiple products.
- Wishlist is designed as an independent entity to allow users to save products without purchasing them immediately.
- Shipment and Payment are stored separately to simplify order tracking and transaction management.

This design improves scalability, consistency, and maintainability of the database.
